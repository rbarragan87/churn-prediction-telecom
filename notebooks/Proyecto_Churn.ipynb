# Análisis del Churn en una Empresa de Telefonía.

**Contexto:**
El churn (abandono de clientes) es un problema crítico para las empresas, ya que la pérdida de clientes representa una pérdida de ingresos significativa. Este análisis busca comprender los factores que impulsan el churn en una empresa de telefonía, para identificar patrones y generar estrategias efectivas de retención.

**Objetivo:**
El propósito de este análisis es:

- Identificar las características de los clientes que abandonan el servicio.
- Cuantificar el impacto financiero del churn, tanto en términos mensuales como anuales.
- Examinar las variables clave que pueden estar asociadas con el abandono de clientes.
- Realizar y contestar preguntas enfocadas en el negocio.

# Definición del problema

**Condiciones de la asignación principal**

Algunos otros servicios que ofrece la empresa incluyen:

La clientela puede elegir entre un pago mensual o firmar un contrato de 1 o 2 años. Puede utilizar varios métodos de pago y recibir una factura electrónica después dtir del 1 de febrero de 2020.

Antes de empezar el análisis más profundo me gustaría hacer algunas modificaciones en mi dataset:

1. Crear variable de antiguedad: Tengo datos con inicios de contrato de 01/10/2013 a 01/02/2020 y con cierres de contrato del 01/10/2019 al 01/01/2020. Debido a que tengo datos hasta el 2020 usaré como fecha actual 01/02/2020.

```python
# Crear variable tipo fecha con la fecha actual que usaré 
fecha_actual = pd.to_datetime('2020-02-01')

# Crear la columna 'RefDate' con la fecha actual
df['RefDate'] = fecha_actual
```

```python
# Usaré esta condición para crear la antiguedad
df['RefDate'] = np.where(
    df['EndDate'].isna(),  # Si en EndDate hay NaT (cliente está activo)
    df['RefDate'],  # Usar fecha_actual
    df['EndDate']  # Si no, usar Fecha_Cierre_Contrato
)
```

```python
# Calcular la antigüedad en días usando timedelta64 y como argumentos paso 1 porque es por día y 'D' para especiifcar que con días
df['Tenure'] = (df['RefDate'] - df['BeginDate']) / np.timedelta64(1, 'D')

# Convertir la antigüedad de días a meses
df['TenureMonths'] = df['Tenure'] / 30
df['TenureMonths'] = df['TenureMonths'].round().astype(int) #Redondeo y convierto a int
```

```python
# Eliminar la columna 'RefDate' ya que le cree de manera temporal
df.drop(columns=['RefDate'], inplace=True)
```

## Feature Importance

Para saber que variables son más relevantes para predecir el churn usaré 2 distintos métodos:

2. Método Boruta

**1. Método feature_importances**

```python
# Hago una copia del df y codifico variables categóricas (para convertirlas en numéricas)
df_encoded = df.copy()
for col in df_encoded.select_dtypes(include=["object", "category"]).columns:
    df_encoded[col] = LabelEncoder().fit_transform(df_encoded[col])

# Separo variables predictoras y objetivo (Churn)
X = df_encoded.drop(columns=["Churn"])
y = df_encoded["Churn"]

# Entrenar modelo Random Forest
model = RandomForestClassifier(random_state=42)
model.fit(X, y)

# Saco la importancia de las variables
importances = pd.Series(model.feature_importances_, index=X.columns)
importances = importances.sort_values(ascending=True)

# Grafiquemos el resultado
plt.figure(figsize=(10, 6))
importances.plot(kind="barh", color="skyblue")
plt.xlabel("Importancia")
plt.ylabel("Variables")
plt.title("Importancia de Variables en la Predicción de Churn")
plt.show()
```

```python
# Creo una instancia BorutaPy con el modelo Random Forest y le paso random_state para que pueda replicarse
ft_select = BorutaPy(model, n_estimators="auto", random_state=42)

# Ajusto el modelo con los datos 
ft_select.fit(X, y)

# Obtener los nombres de las columnas que fueron seleccionadas como importantes por BorutaPy
ft_importance = X.columns[ft_select.support_]
ft_importance
```

**4. Train-Test Split**

Consideraciones antes de separar el Train y Test:
- Regresión Logística: Es sensible a la escala de las variables, por lo que la estandarización es importante.
- Algoritmos basados en arboles como; Random Forest, Decision Tree y XGBoost: No requieren estandarización de las variables, pero pueden beneficiarse de la codificación de variables categóricas.

```python
# Datos con solo las categóricas procesadas (ideales para RandomForest y XGBoost)
X_train_raw, X_test_raw, y_train_raw, y_test_raw = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
# Datos con variables categóricas y numpericas transformadas (Ideal para Regresión)
X_train_processed, X_test_processed, y_train_processed, y_test_processed = train_test_split(X_encoded_and_standardized, y, test_size=0.2, random_state=42)
```

**5. Desbalanceo de Clases**

Usaré el submuestreo SMOTE, pero solamente en los datos que le pasaré a la Regresión Logística, Decision Tree y Random Forest, ya que XGBoost puede trabajar con desbalanceo de clases.

```python
# Ver la distribución antes de aplicar SMOTE
y_train_raw.value_counts()
```

```python
# Llamo a SMOTE con random_state
smote = SMOTE(random_state=42)

# Aplicar SMOTE al conjunto de entrenamiento sin mucho proceso
X_train_raw_res, y_train_raw_res = smote.fit_resample(X_train_raw, y_train_raw)

# Aplicar SMOTE al conjunto de entrenamiento completamente procesado
X_train_processed_res, y_train_processed_res = smote.fit_resample(X_train_processed, y_train_processed)

# Validar que se haya aplicado correctamente la distribución de clases en los nuevos df 'resampleados'
print("Distribución de clases en el conjunto de entrenamiento sin mucho proceso (después de SMOTE):")
print(y_train_raw_res.value_counts())
print()
print("Distribución de clases en el conjunto de entrenamiento completamente procesado (después de SMOTE):")
print(y_train_processed_res.value_counts())
```

        - Codificación (pd.get_dummies)

# 7. Modelado

### Elección de Modelos

- Para este proyecto seleccioné una combinación de modelos tanto lineales como no lineales, con el objetivo de evaluar su rendimiento en un problema de clasificación binaria con clases desbalanceadas.
- Utilice GridSearchCV en el modelado para realizar Cross Validation y buscar el mejor rendimiento de cada modelo eligiendo distintos hiperparámetros.

---
**Regresión Logística (LogisticRegression):** 

Es un modelo base simple pero efectivo, fácil de interpretar. Sirve como punto de referencia para comparar otros modelos más complejos. Además, su interpretabilidad ayuda a entender qué variables influyen más en el churn.

- Datos usados: X_train_processed_res, y_train_processed_res.
- Justificación:
    - La regresión logística es sensible a la escala de los datos, por eso es necesario estandarizar.
    - No maneja el desbalance de clases internamente, por eso se aplicó SMOTE.

---

**Árbol de Decisión (DecisionTreeClassifier):** 

Permite visualizar la lógica detrás de las decisiones y es útil para capturar relaciones no lineales. Aunque puede sobreajustarse, es un buen modelo base para entender estructuras de decisión.

- Datos usados: X_train_raw_res, y_train_raw_res
- Justificación:
    - Los árboles no requieren estandarización.
    - Se usó SMOTE porque el modelo no compensa internamente el desbalance.

---
**Bosque Aleatorio (RandomForestClassifier):** 

Es una técnica de ensamblado que mejora la precisión y estabilidad al combinar múltiples árboles. Es más robusto frente al sobreajuste y puede manejar bien datos desbalanceados si se ajustan los pesos o se usan técnicas de balanceo.

- Datos usados: X_train_raw_res, y_train_raw_res
- Justificación:
    - Igual que el árbol de decisión, no necesita escalar datos.
    - Requiere balanceo externo como SMOTE si las clases están desbalanceadas.

---
**XGBoost (XGBClassifier):** 

Es uno de los modelos más potentes en problemas estructurados/tabulares. Permite un mejor manejo del desbalance usando el parámetro scale_pos_weight, que ajusté en función de la proporción entre clases. Se incluye como modelo avanzado debido a su rendimiento probado en muchos problemas de clasificación binaria con desequilibrio.

- Datos usados: X_train_raw, y_train_raw
- Justificación:
    - XGBoost ya incluye manejo interno del desbalance con el parámetro scale_pos_weight, por lo que no se necesita aplicar SMOTE.
    - No requiere estandarización porque los árboles que usa no se ven afectados por la escala.

---

**Uso de Cross Validation y Grid Search**

Utilicé GridSearchCV con validación cruzada (CV=5) para encontrar los mejores hiperparámetros para cada modelo, evitando sobreajuste a los datos de entrenamiento. La validación cruzada asegura que el modelo sea evaluado en distintos subconjuntos de los datos y que el rendimiento general sea más estable y representativo.

Aunque la métrica usada para la búsqueda de hiperparámetros fue inicialmente accuracy, al final evalué los modelos con métricas más adecuadas al problema, dado el desbalance de clases.

**Función para evaluar los modelos**

```python
# Crear listas vacías con las métricas de evaluación
accuracy = []
recall = []
roc_auc = []
precision = []
f1 = []

# Función para calcular y mostrar métricas
def calcular_mostrar_metricas(y_test, y_pred, model_name):
    acc = round(accuracy_score(y_test, y_pred), 5)
    rec = round(recall_score(y_test, y_pred), 5)
    roc = round(roc_auc_score(y_test, y_pred), 5)
    prec = round(precision_score(y_test, y_pred), 5)
    f1_score_val = round(f1_score(y_test, y_pred), 5)
    # Imprimir las métricas de cada modelo
    print(f"{model_name} - Accuracy: {acc}, Recall: {rec}, Roc_Auc: {roc}, Precision: {prec}, F1: {f1_score_val}")
    # Guardar las métricas en la lista para mostrar al final
    accuracy.append(acc)
    recall.append(rec)
    roc_auc.append(roc)
    precision.append(prec)
    f1.append(f1_score_val)
```

**Llamar a los modelos indicando sus hiperparámetros e inicializar GridSearchCV**

```python
# Llamo a los modelos a entrenar
logistic_regression = LogisticRegression(random_state=42, max_iter=1000)
decision_tree = DecisionTreeClassifier(random_state=42)
random_forest = RandomForestClassifier(random_state=42)
xgbc_base = xgb.XGBClassifier(random_state=0, scale_pos_weight=3)

# Definir hiperparámetros
param_grid_lr = {
    'C': [0.1, 10, 20, 100],
    'solver': ['liblinear', 'lbfgs']}

param_grid_dt = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]}
    
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]}

param_grid_xgb = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 10],
    'learning_rate': [0.01, 0.1, 0.3],
    'scale_pos_weight': [1, 3, 5]}

# Inicializar GridSearchCV para cada modelo
grid_search_lr = GridSearchCV(estimator=logistic_regression, param_grid=param_grid_lr, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_dt = GridSearchCV(estimator=decision_tree, param_grid=param_grid_dt, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_rf = GridSearchCV(estimator=random_forest, param_grid=param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_xgb = GridSearchCV(estimator=xgbc_base, param_grid=param_grid_xgb, cv=5, scoring='accuracy', n_jobs=-1)
```

**Ajusto los modelos**

```python
# Ajustar los modelos en los datos de entrenamiento correspondientes que favorecen a cada modelo
grid_search_lr.fit(X_train_processed_res, y_train_processed_res)
grid_search_dt.fit(X_train_raw_res, y_train_raw_res)
grid_search_rf.fit(X_train_raw_res, y_train_raw_res)
grid_search_xgb.fit(X_train_raw, y_train_raw)
```

# 8. Evaluación y selección del modelo final

**Métricas de Evaluación**

- Recall (Sensibilidad): Mide cuántos de los verdaderos clientes que cancelaron fueron correctamente identificados. Esta métrica es clave en este contexto, ya que nos interesa capturar la mayor cantidad posible de clientes que abandonan.
- Precision: Mide cuántos de los clientes que el modelo predijo que cancelarían realmente lo hicieron.
- F1 Score: Es el balance entre precision y recall.
- ROC AUC: Evalúa la capacidad del modelo para separar ambas clases (churn vs no churn). Es insensible al desbalance de clases y permite comparar modelos de manera más objetiva.
- Accuracy: Se mantiene solo como referencia para tener un panorama general.

**NOTA: Debido a que el objetivo de mi modelo es predecir el Churn tomaré como méticra principal para determinar el modelo ganador al Recall, respaldado por ROC AUC y F1.**

**Evaluación de los modelos**

```python
# Evaluar los modelos optimizados
y_pred_lr = grid_search_lr.predict(X_test_processed)
y_pred_dt = grid_search_dt.predict(X_test_raw)
y_pred_rf = grid_search_rf.predict(X_test_raw)
y_pred_xgb = grid_search_xgb.predict(X_test_raw)

# Mostrar los mejores parámetros y la mejor precisión obtenida
print(f"Best Logistic Regression Accuracy: {grid_search_lr.best_score_:.4f}")
print(f"Best Decision Tree: {grid_search_dt.best_score_:.4f}")
print(f"Best Random Forest Accuracy: {grid_search_rf.best_score_:.4f}")
print(f"Best XGBoost Accuracy: {grid_search_xgb.best_score_:.4f}")
print()
print(f"Best Logistic Regression Params: {grid_search_lr.best_params_}")
print(f"Best Decision Tree: {grid_search_dt.best_params_}")
print(f"Best Random Forest Params: {grid_search_rf.best_params_}")
print(f"Best XGBoost Params: {grid_search_xgb.best_params_}")
```

```python
# Llamar a la función para visualizar las otra métricas de evaluación
calcular_mostrar_metricas(y_test_processed, y_pred_lr, "Logistic Regression Optimizado")
calcular_mostrar_metricas(y_test_raw, y_pred_rf, "Random Forest Optimizado")
calcular_mostrar_metricas(y_test_raw, y_pred_dt, "Decision Tree Optimizado")
calcular_mostrar_metricas(y_test_raw, y_pred_xgb, "XGBoost Optimizado")
```

Resumen:
